```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
df = table[0]
df
```

- Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df = replace_
columns = list(df.columns)
c = pd.DataFrame(columns, columns=['Колонка'])
res=list(map(lambda x: f'{(df[x].isnull().sum() * 100 / len(df[x])).round(2)}%', df.columns))
new = pd.concat([pd.DataFrame(list(df.columns), columns=['Колонка']), pd.DataFrame(res, columns=['Частка пропусків,%'])], axis=1)
new.style.set_table_styles(
   [{
       'selector': 'thead',
       'props': [('background-color', '#add8e6')]
   }]).set_properties(**{'text-align': 'center'}).hide(axis="index")


```

- Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = replace_
for col in df.columns:
  df[col] = pd.to_numeric(df[col], errors='ignore')
df = df.fillna({i: df[i].mean().round(2) for i in df.columns[1:]})
df

```

- Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
a = df['2019'][27]
print(f'Значення більші за {a}')
res = df[df['2019'] > a]
res[['Регіон', '2019']]
```

```python
data = res['2019']
labels = res['Регіон']
plt.pie(data, labels=labels,  autopct="%.2f%%")
plt.show()
```

- Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.figure(figsize=(10,6))
plt.bar(df['Регіон'].loc[0:26], df['2019'].loc[0:26], color=['cadetblue'])
plt.title('Діаграма народжуваності по регіонах у 2019 році')
plt.xlabel('Регіони')
plt.ylabel('Коефіціент народжуваності')
plt.xticks(rotation=90)
plt.legend(['2019 рік'])
plt.show()
```
