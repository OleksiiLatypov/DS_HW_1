- Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
for column in data.columns:
  if data[column].isnull().sum() and column != 'Язык.программирования':
    data = data.drop(column, axis=1)
data.head()
```

- Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary = python_data.groupby(by='Должность')['Зарплата.в.месяц'].agg(['min', 'max'])
salary
```

- Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
# salary['avg'] = salary[['min','max']].mean(axis=1)
# salary

# salary['avg'] = salary.apply(lambda x: x.mean(), axis=1)
# salary

def fill_avg_salary(salary):
  return salary.mean()

salary['avg'] = salary.apply(fill_avg_salary, axis=1)
salary
```

- Graphik

```python

salary.sort_values(by='avg', ascending=False)
position = list(salary.index)

salaries = list(salary['avg'].sort_values(ascending=True))

salaries_1 = list(salary['min'].sort_values(ascending=True))

salaries_2 = list(salary['max'].sort_values(ascending=True))

df_plot = pd.DataFrame({'avg': salaries, 'min': salaries_1, 'max': salaries_2}, index=position)
df_plot.plot.barh(color=['lightsteelblue', 'g', 'orange'])
plt.title('Salary')
plt.xlabel('Зарплата, $', fontsize='small', color='midnightblue')
plt.ylabel('Посада', fontsize='small', color='midnightblue')
plt.show()
```
